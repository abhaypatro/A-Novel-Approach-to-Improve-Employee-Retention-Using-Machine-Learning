```python
train = h2o.import_file("adasyn.csv")
```

```python
train
```

```python
x = train.columns
y = "Attrition"
x.remove(y)
```

```python
train[y] = train[y].asfactor()
```

```python
aml = H2OAutoML(max_models=40, seed=1,verbosity="info")
aml.train(x=x, y=y, training_frame=train)
```

```python
lb = aml.leaderboard
lb.head(rows=lb.nrows)
```

```python
test= h2o.import_file("/content/testadasyn.csv")
```

```python
test[y] = test[y].asfactor()
```

```python
# Ensembled model 
a=h2o.get_model('StackedEnsemble_AllModels_AutoML_20210329_082903')
```

```python
pr=a.predict(test)
data_as_df = pr.as_data_frame()
```

```python
y_pred=data_as_df.predict
y_test=pd.read_csv('/content/testadasyn.csv').Attrition
```

```python
pr
```

```python
from sklearn.metrics import f1_score
f1_score(y_test,y_pred)
```
