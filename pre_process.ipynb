```python
for i in range(len(train['BusinessTravel'])):
  if train['BusinessTravel'][i]=='Non-Travel':
    train['BusinessTravel'][i]=1
  elif train['BusinessTravel'][i]=='Travel_Rarely':
    train['BusinessTravel'][i]=2
  else:
    train['BusinessTravel'][i]=3
```

```python
train
```

```python
for i in range(len(train['Gender'])):
  if train['Gender'][i]=='Male':
    train['Gender'][i]=1
  else:
    train['Gender'][i]=0
```

```python
#Label encoding done on Buisness Travel,gender,OverTime,maritalstatus
# 1-> Male, 0->Female
# 1->No_Travel, 2->Travel_rarely, 3->Frequent_travel
# 1->Yes overtime, 0->No overtime
# 1->single,2->divorced,3->married
```

```python
for i in range(len(train['Attrition'])):
  if train['Attrition'][i]=='Yes':
    train['Attrition'][i]=1
  else:
    train['Attrition'][i]=0
```

```python
for i in range(len(train['OverTime'])):
  if train['OverTime'][i]=='Yes':
    train['OverTime'][i]=1
  else:
    train['OverTime'][i]=0
```

```python
for i in range(len(train['MaritalStatus'])):
  if train['MaritalStatus'][i]=='Single':
    train['MaritalStatus'][i]=1
  elif train['MaritalStatus'][i]=='Divorced':
    train['MaritalStatus'][i]=2
  else:
    train['MaritalStatus'][i]=3
```

```python
# One-Hot encoding to be done on Department,EducationField,JobRole.
```
