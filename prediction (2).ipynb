```python
fulllis=[]
for a in range(len(slack)):
  global fulllis
  lis=[]
  if slack[a]=='MonthlyIncome':
      for i in range(10):
        lis.append(int(t[slack[a]][2:3])+i*100)
  elif slack[a]=='YearsInCurrentRole' or slack[a]=='YearsSinceLastPromotion' or slack[a]=='YearsWithCurrManager':
        lis.append(0)
  else:
      for i in range(min(train[slack[a]]),max(train[slack[a]])+1):
        lis.append(i)
  fulllis.append(lis)
```

```python
fulllis
```

```python
import itertools
gcounter=1
for x in list(itertools.product(*fulllis)):
  temp=t[2:3].copy()
  for i in range(len(x)):
    temp[slack[i]][0:1]=x[i]
  pr=model.predict(h2o.H2OFrame(temp))
  print(gcounter,"--->",int(pr.as_data_frame().predict))
  gcounter=gcounter+1
```
